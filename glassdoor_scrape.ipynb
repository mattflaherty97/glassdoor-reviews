```python
# append everything captured on a page to this df
# reviews = pd.DataFrame()
reviews = pd.read_csv('C:/Users/19012/Documents/sp22-capstone/company_reviews.csv')

# open web browser
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path="/Users/19012/Documents/chromedriver", options=options)

# go to Glassdoor to sign in
# url = 'https://www.glassdoor.com/Search/results.htm?keyword=' + company_name
url = 'https://www.glassdoor.com/index.htm'
driver.get(url)

time.sleep(5)

try:
    # sign in to Glassdoor
    # sign_in = driver.find_elements_by_xpath('//*[@id="SignInButton"]')[1]
    sign_in = driver.find_elements_by_xpath('//*[@id="SiteNav"]/nav/div[2]/div/div/div/button')[0]
    # sign_in = driver.find_elements_by_css_selector(".d-flex.align-items-center.justify-content-center.order-1.order-md-2.mr-auto.mr-md-0.p-0.LockedHomeHeaderStyles__signInButton")[0]
    sign_in.click()
except:
    pass
    
time.sleep(2)

# sometimes an ad will pop up which won't allow you to enter information
try:
    click_x = driver.find_elements_by_css_selector('.qual_x_close')[0]
    click_x.click()
    
    time.sleep(2)
    
    # sign in to Glassdoor
    # sign_in = driver.find_elements_by_xpath('//*[@id="SignInButton"]')[1]
    sign_in = driver.find_elements_by_xpath('//*[@id="SiteNav"]/nav/div[2]/div/div/div/button')[0]
    # sign_in = driver.find_elements_by_css_selector(".d-flex.align-items-center.justify-content-center.order-1.order-md-2.mr-auto.mr-md-0.p-0.LockedHomeHeaderStyles__signInButton")[0]
    sign_in.click()
except:
    pass
    
try:
    # login information
    email_address = driver.find_elements_by_xpath('//*[@id="modalUserEmail"]')[0]
    email_address.send_keys("matt.flaherty97@gmail.com")

    # password
    password = driver.find_elements_by_xpath('//*[@id="modalUserPassword"]')[0]
    password.send_keys("Gotigers1")

    # click sign in button to officially sign in
    login_button = driver.find_elements_by_css_selector(".gd-ui-button.mt-std.minWidthBtn.css-14xfqow.evpplnh0")[0]
    login_button.click()
except:
    pass

time.sleep(5)
    
# loop through companies
# trouble companies: MEREDITH CORP, E TRADE FINANCIAL CORP
for company_name in fortune_500[fortune_500.index('NEWMONT MINING CORP'):]:
    
    url = 'https://www.google.com/search?q=' + company_name + ' glassdoor reviews'
    try:
        # search for company on google
        driver.get(url)
    except:
        driver.refresh()
        
    time.sleep(5)
    
    # make sure the link we are about to click is a glassdoor link
    hyperlinks = driver.find_elements_by_css_selector('.yuRUbf')
    for i in range(len(hyperlinks)):
        try:
            object_detect = hyperlinks[i].find_elements_by_tag_name('a')
            final = object_detect[0].get_attribute('href')
            if ('glassdoor' in final) and ('Reviews' in final):
                break
        except:
            pass
        
    # click link that takes us to Glassdoor reviews
    try:
        company = driver.find_elements_by_css_selector(".LC20lb.MBeuO.DKV0Md")
        company[i].click()
    except:
        driver.refresh()
    
    time.sleep(5)
    
    # if there are fewer than 10 reviews (number above 'review' <= 10), move on to next item
    # the scraping is thrown off for these companies and it's only 10 reviews
    try:
        review_num_top = driver.find_elements_by_css_selector('.count')
        review_num_top = review_num_top[0].text.split('K')
        if (int(review_num_top[0]) <= 10) & (len(review_num_top) == 1):
            continue
            
        time.sleep(5)
    except:
        pass
    
    # workaround
    try:
        workaround = driver.find_elements_by_css_selector('.px-lg-0.px')
        if 'Showing results for ' in workaround[0].text:
            continue
    except:
        pass
    
    # some pages require you to click 'see all reviews'
    # for those that don't, just move on to the next part of the script
    try:
        see_all_reviews = driver.find_elements_by_xpath('//*[@id="NodeReplace"]/main/div[2]/div[1]/div[1]/div/div[2]/strong/span[1]/a')
        see_all_reviews[0].click()
        
        time.sleep(5)
    except:
        pass
    
    try:
        # click the dropdown to choose what to sort the reviews by
        driver.find_elements_by_css_selector(".css-m3gjah.egu3u860")[0].click()
        
        time.sleep(5)
    
        # click on sort by most recent
        driver.find_elements_by_xpath('//*[@id="option_DATE"]/span')[0].click()
    except:
        driver.refresh()
    
    time.sleep(5)
    
    # turns to false when there are no more pages of reviews
    more_pages = True

    # run until more_pages = False
    while more_pages:
        reviews_list = []

        # some reviews have a 'Continue reading' option, we need to click this button to scrape the rest of the text
        # open them all before scraping
        continue_reading = driver.find_elements_by_css_selector(".v2__EIReviewDetailsV2__continueReading.v2__EIReviewDetailsV2__clickable.v2__EIReviewDetailsV2__newUiCta.mb")
        for reading in range(len(continue_reading)):
            try:
                continue_reading[reading].click()
                time.sleep(2)
                # if something pops up
                driver.find_elements_by_xpath('//*[@id="qual_ol"]/div[1]')[0].click()
                #time.sleep(5)
            except:
                continue

        # find the cards with 'Advice to Management'
        with_advice = []
        everything = driver.find_elements_by_css_selector('.noBorder.empReview.cf.pb-0.mb-0')
        for item in range(len(everything)):
            try:
                with_advice.append(everything[item].text)
            except StaleElementReferenceException:
                continue

        for advice in range(len(with_advice)):
            if 'Advice to Management' in with_advice[advice]:
                with_advice[advice] = 1
            else:
                with_advice[advice] = 0
                
        def get_text(web_element):
            try:
                return web_element.text
            except StaleElementReferenceException:
                pass
    
        # employee rating
        ratings = driver.find_elements_by_xpath('//span[@class="ratingNumber mr-xsm"]')
        result = map(get_text, ratings)
        ratings_list = list(result)        
        
        #ratings_list = []
        #for rating in range(len(ratings)):
        #    try:
        #        ratings_list.append(ratings[rating].text)
        #    except StaleElementReferenceException:
        #        continue
        
        ################# subratings
        # number of stars classes
        v = ['css-xd4dom e1hd5jg10', 'css-18v8tui e1hd5jg10', 'css-vl2edp e1hd5jg10', 'css-1nuumx7 e1hd5jg10', 'css-s88v13 e1hd5jg10']
        
        def sub_rating(sub_num):
            r = driver.find_elements_by_css_selector('.gdReview')
            sub_list = []
            for i in range(len(r)):
                try:
                    if len(r[i].find_elements_by_css_selector('.tooltipContainer')) != 0:
                        srdiv = r[i].find_elements_by_tag_name('li')
                        srclass = srdiv[sub_num].find_elements_by_tag_name('div')[1].get_attribute('class')
                        sub_list.append(v.index(srclass)+1)
                    else:
                        sub_list.append(np.nan)
                except:
                    sub_list.append(np.nan)
            return sub_list
                
        # work/life balance
        work_life_list = sub_rating(0)
        
        # culture & values
        culture_values_list = sub_rating(1)
        
        # diversity & inclusion
        diverse_include_list = sub_rating(2)
        
        # career opportunities
        career_opps_list = sub_rating(3)
        
        # compensation and benefits
        comp_benefits_list = sub_rating(4)
        
        # senior management
        senior_manage_list = sub_rating(5)
        
        # employee status (current/former)
        employment_status = driver.find_elements_by_css_selector('.pt-xsm.pt-md-0.css-1qxtz39.eg4psks0')
        result = map(get_text, employment_status)
        status_list = list(result)
        
        #status_list = []
        #for status in range(len(employment_status)):
        #    try:
        #        status_list.append(employment_status[status].text)
        #    except StaleElementReferenceException:
        #        continue
        
        # employee info
        author_info = driver.find_elements_by_xpath('//span[@class="authorInfo"]')
        result = map(get_text, author_info)
        info_list = list(result)
        
        #info_list = []
        #for info in range(len(author_info)):
        #    try:
        #        info_list.append(author_info[info].text)
        #    except StaleElementReferenceException:
        #        continue
        
        #################### Recommend, CEO approval, and Business Outlook
        # the second part after the hyphen represents check, circle, line, or X
        # check == 1
        # circle == 2
        # neutral == 3
        # X == 4
        # [check, circle, line, X]
        marks = ['SVGInline css-hcqxoa d-flex', 'SVGInline css-10xv9lv d-flex', 'SVGInline css-1h93d4v d-flex', 'SVGInline css-1kiw93k d-flex']

        objects = driver.find_elements_by_css_selector('.d-flex.align-items-center.mr-std')
        marks_list = []
        for i in range(len(objects)):
            try:
                object_detect = objects[i].find_elements_by_tag_name('span')
                final = object_detect[0].get_attribute('class')
                marks_list.append(marks.index(final)+1)
            except:
                pass
           
        # recommend
        try:
            recommend = np.array(marks_list).reshape((10,3))[:,0]
        except:
            pass
        
        # CEO approval 
        try:
            ceo_approval = np.array(marks_list).reshape((10,3))[:,1]
        except:
            pass
        
        # Business Outlook
        try:
            outlook = np.array(marks_list).reshape((10,3))[:,2]
        except:
            pass
        
        # pros
        pros = driver.find_elements_by_xpath('//span[@data-test="pros"]')
        result = map(get_text, pros)
        pros_list = list(result)
        
        #pros_list = []
        #for pro in range(len(pros)):
        #    try:
        #        pros_list.append(pros[pro].text)
        #    except StaleElementReferenceException:
        #        continue

        # cons
        cons = driver.find_elements_by_xpath('//span[@data-test="cons"]')
        result = map(get_text, cons)
        cons_list = list(result)
        
        #cons_list = []
        #for con in range(len(cons)):
        #    try:
        #        cons_list.append(cons[con].text)
        #    except StaleElementReferenceException:
        #        continue

        # advice to management
        advices = driver.find_elements_by_xpath('//span[@data-test="advice-management"]')
        result = map(get_text, advices)
        advice_list = list(result)
        
        #advice_list = []
        #for advice in range(len(advices)):
        #    try:
        #        advice_list.append(advices[advice].text)
        #    except StaleElementReferenceException:
        #        continue

        for i in range(len(with_advice)):
            for j in range(len(advice_list)):
                if with_advice[i] == 1:
                    with_advice[i] = advice_list[0]
                    advice_list.pop(0)
        try:
            reviews_list = {"company": [company_name] * len(ratings_list),
                            "rating" : ratings_list,
                            "sub_work_life_balance": work_life_list,
                            "sub_culture_values": culture_values_list,
                            "sub_diversity_inclusion": diverse_include_list,
                            "sub_career_opportunities": career_opps_list,
                            "sub_compensation_benefits": comp_benefits_list,
                            "sub_senior_management": senior_manage_list,
                            "author_info" : info_list,
                            "employee_status": status_list,
                            "recommend": recommend,
                            "ceo_approval": ceo_approval,
                            "outlook": outlook,
                            "pros" : pros_list,
                            "cons" : cons_list,
                            "advice_management" : with_advice}
        except:
            pass
        
        # if the date in author info is < November 2018, move on to the next company
        # this will give us 2021 and 2 months in 2022 (post-pandemic) and 2019 and 2 months in 2018 (pre-pandemic)
        # we will also have all of 2020 (pandemic)

        # add the current page's reviews to the table
        # refresh and move on to the next page if the lists are not the same size which causes an error
        # there are thousands of reviews for these companies, so we can live without 10 reviews
        try:
            reviews = reviews.append(pd.DataFrame(reviews_list), ignore_index=True)
            print('reviews scraped:', reviews.shape[0])
        except ValueError:
            driver.refresh()
            time.sleep(2)
        
        # go to next review if max number of reviews is reached or we have scraped through 2018
        try:
            new_list = []
            final_review = driver.find_elements_by_css_selector('.paginationFooter')
            for i in final_review[0].text.split(' '):
                try:
                    int(i.replace(',', ''))
                    new_list.append(int(i.replace(',', ''))) 
                except:
                    pass

            if (new_list[1] == new_list[2]) | (int(reviews.author_info.str.split(' ').iloc[-1][2]) < 2018):
                more_pages = False
        except:
            try:
                new_list = []
                final_review = driver.find_elements_by_css_selector('.d-flex.justify-content-center.css-1dcha3m.e5zbkfc0')
                for i in final_review[0].text.split(' '):
                    try:
                        int(i.replace(',', ''))
                        new_list.append(int(i.replace(',', ''))) 
                    except:
                        pass

                if (new_list[1] == new_list[2]) | (int(reviews.author_info.str.split(' ').iloc[-1][2]) < 2018):
                    more_pages = False
            except:
                driver.refresh()
        
        # sometimes the page won't load, refresh the page in this instance
        try:
            # next page
            next_button = driver.find_elements_by_css_selector('.nextButton.css-1hq9k8.e13qs2071')[0]
            # clicking on the button
            next_button.click()
        # common errors that I've gotten: ElementClickInterceptedException (pop-up window), IndexError (page did not load properly)
        # I think the best handling is just to refresh the page for any error to remove pop up windows
        # this method will also allow the page to load properly again
        except IndexError:
            driver.refresh()
        except ElementClickInterceptedException:
            driver.refresh()
            
        time.sleep(2)

# close browser
driver.quit()

# reviews
```
